```python
train = pd.read_csv("https://raw.githubusercontent.com/fkvn/Toxic-Comments/master/Files/train.csv")
```

```python
# list of feature names that only contain numerical:
feature_cols = ['comment_text']

# use the above list to select the features from the original DataFrame
X_train = train[feature_cols]  

# print the first 5 rows
X_train.head()
```

```python
# checking the size of Feature Matix X:

print(X_train.shape)
```

```python
# Select a Series of labels
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
y_train = train[label_cols]

# checking the label vector by printing every 10 values
y_train[::10]
```

```python
test = pd.read_csv("https://raw.githubusercontent.com/fkvn/Toxic-Comments/master/Files/test.csv")

test[0::10]
```

```python

# list of feature names that only contain numerical:
feature_cols = ['comment_text']

# use the above list to select the features from the original DataFrame
X_test = test[feature_cols]  

# print the first 5 rows
X_test.head()
```

```python
test_labels = pd.read_csv("https://raw.githubusercontent.com/fkvn/Toxic-Comments/master/Files/test_labels.csv")

test_labels[0::10]
```

```python
# Select a Series of labels
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
y_test = test_labels[label_cols]

# checking the label vector by printing every 10 values
y_test[::10]
```

```python
# Step1: Import KNeighborsClassifier "Class"
from sklearn.neighbors import KNeighborsClassifier

# import the accuracy 
from sklearn.metrics import accuracy_score

# import the LabelEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
# import the LabelEncoder
from sklearn.preprocessing import LabelEncoder

## label encode the categorical values and convert them to numbers 
le = LabelEncoder()
X_train['comment_text'] = le.fit_transform(X_train['comment_text'])

le = LabelEncoder()
X_test['comment_text'] = le.fit_transform(X_test['comment_text'])


```

```python
# X_test
```

```python
# # Step2: Instantiate a KNN object with k = 3
k = 3
my_knn_for_toxic_comments = KNeighborsClassifier(n_neighbors=k) # name of the object is arbitrary!

# Step3: Training ONLY the training set
my_knn_for_toxic_comments.fit(X_train, y_train)

# Step4: Testing on the testing set
y_predict = my_knn_for_toxic_comments.predict(X_test)


# # # Step5: Evaluating the accuracy 
accuracy =  accuracy_score(y_test, y_predict)

print("Accuracy KNN: ", accuracy)

```
