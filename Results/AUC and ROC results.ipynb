# This file will calculate AUC and ROC of 6 methods: 

* **OHE Logistic Regression**
* **OHE Randomforest**
* **TFIDF Logistic Regestion**
* **Word2vec Decision Tree**
* **Word2vec Logistic Regestion**
* **Word2vec Randomforest**

### with 153165 test samples 

```python
import pandas as pd
import numpy as np

# Importing the "pyplot" package of "matplotlib" library of python to generate 
# graphs and plot curves:
import matplotlib.pyplot as plt


label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

test_labels = pd.read_csv('test_labels.csv')
test_labels[label_cols] = test_labels[label_cols].replace(-1, 1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

OHELogReg_Labels = pd.read_csv('OHE_LogReg_submission.csv')
OHERandomforest_Labels = pd.read_csv('OHE_Randomforest_submission.csv')
TFIDFLogReg_Labels = pd.read_csv('TFIDF_LogReg_submission.csv')
W2VDectree_Labels = pd.read_csv('Word2vec_DecTree_submission.csv')
W2VLogReg_Labels = pd.read_csv('Word2vec_LogisticReg_submission.csv')
W2VRandomForest_Labels = pd.read_csv('Word2vec_Randomforest_submission.csv')


for col in label_cols:
    print("\n**", col, "**: ")
    y_test = test_labels[col]
    
    # y_prediction    
    y_pred_OHELogReg = OHELogReg_Labels[col]
    y_pred_OHERandomforest = OHERandomforest_Labels[col]
    y_pred_TFIDFLogReg = TFIDFLogReg_Labels[col]
    y_pred_W2VDectree = W2VDectree_Labels[col]
    y_pred_W2VLogReg = W2VLogReg_Labels[col]
    y_pred_W2VRandomForest = W2VRandomForest_Labels[col]
    
    
    # OHELogReg AUC    
    fprOHELogReg, tprOHELogReg, thresholdsOHELogReg = metrics.roc_curve(y_test, y_pred_OHELogReg, pos_label=1)
    AUCOHELogReg = metrics.auc(fprOHELogReg, tprOHELogReg)
    print("OHELogReg - AUC: ", AUCOHELogReg)
    
    # OHERandomforest AUC
    fprOHERandomforest, tprOHERandomforest, thresholdsOHERandomforest = metrics.roc_curve(y_test, y_pred_OHERandomforest, pos_label=1)
    AUCOHERandomforest = metrics.auc(fprOHERandomforest, tprOHERandomforest)
    print("OHERandomforest - AUC: ", AUCOHERandomforest)
    
    # OHELogRegTFIDFLogReg AUC
    fprTFIDFLogReg, tprTFIDFLogReg, thresholdsTFIDFLogReg = metrics.roc_curve(y_test, y_pred_TFIDFLogReg, pos_label=1)
    AUCTFIDFLogReg = metrics.auc(fprTFIDFLogReg, tprTFIDFLogReg)
    print("TFIDFLogReg - AUC: ", AUCTFIDFLogReg)
    
    # W2VDectree AUC
    fprW2VDectree, tprW2VDectree, thresholdsW2VDectree = metrics.roc_curve(y_test, y_pred_W2VDectree, pos_label=1)
    AUCW2VDectree = metrics.auc(fprW2VDectree, tprW2VDectree)
    print("W2VDectree - AUC: ", AUCW2VDectree)
    
    # W2VLogReg AUC
    fprW2VLogReg, tprW2VLogReg, thresholdsW2VLogReg = metrics.roc_curve(y_test, y_pred_W2VLogReg, pos_label=1)
    AUCW2VLogReg = metrics.auc(fprW2VLogReg, tprW2VLogReg)
    print("W2VLogReg - AUC: ", AUCW2VLogReg)
    
    # W2VRandomForest AUC     
    fprW2VRandomForest, tprW2VRandomForest, thresholdsW2VRandomForest = metrics.roc_curve(y_test, y_pred_W2VRandomForest, pos_label=1)
    AUCW2VRandomForest = metrics.auc(fprW2VRandomForest, tprW2VRandomForest)
    print("W2VRandomForest - AUC: ", AUCW2VRandomForest)
    
    # Drawing plot for each class (label) for all 6 methods     
    %matplotlib inline
    plt.figure()
    
    plt.plot(fprOHELogReg, tprOHELogReg, color='yellow', lw=2, 
             label='ROC Curve OHELogReg (area = %0.2f)' % AUCOHELogReg)
    
    plt.plot(fprOHERandomforest, tprOHERandomforest, color='green', lw=2, 
             label='ROC Curve OHERandomforest (area = %0.2f)' % AUCOHERandomforest)
    
    plt.plot(fprTFIDFLogReg, tprTFIDFLogReg, color='red', lw=2, 
             label='ROC Curve TFIDFLogReg (area = %0.2f)' % AUCTFIDFLogReg)
    
    plt.plot(fprW2VDectree, tprW2VDectree, color='magenta', lw=2, 
             label='ROC Curve W2VDectree (area = %0.2f)' % AUCW2VDectree)
    
    plt.plot(fprW2VLogReg, tprW2VLogReg, color='cyan', lw=2, 
             label='ROC Curve W2VLogReg (area = %0.2f)' % AUCW2VLogReg)
    
    plt.plot(fprW2VRandomForest, tprW2VRandomForest, color='black', lw=2, 
             label='ROC Curve W2VRandomForest (area = %0.2f)' % AUCW2VRandomForest)
    
    # Random Guess line:
    plt.plot([0, 1], [0, 1], color='blue', lw=1, linestyle='--')

    # Defining The Range of X-Axis and Y-Axis:
    plt.xlim([-0.005, 1.005])
    plt.ylim([0.0, 1.01])

    # Labels, Title, Legend:
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")

    plt.show()
    

```
